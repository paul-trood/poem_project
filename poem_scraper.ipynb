```python
# import the beautifulsoup library for html parsing
from bs4 import BeautifulSoup
```

```python
# opens the exported kindle notes from the html file
with open("vampire_notebook.html") as f:
    doc = BeautifulSoup(f, "html.parser")
```

```python
tag = doc.title
```

```python
# the words from the kindle book needed are located under the class "noteText"
tags = doc.find_all("div", class_="noteText")
```

```python
# Show tags in the html file
tags
```

```python
# This grabs the text items for all the p tags in the div class
new_tags = [p.get_text() for p in tags]
```

```python
new_tags
```

```python
# import regex library
# Regex will help clean up some of the list items 
import re
```

```python
# There's a list of lists of all items within the div tags
# I only want the word that I highlighted in the book
# The highlighted word is at the beginning of each list

# Generate empty list
result_list = []

# For each item in the new tags list
for i in new_tags:

    # Split each word
    n = i.split()

    # Add each split word to the list generated above
    result_list.append(n)
result_list
```

```python
type(result_list)
```

```python
# Create an entire new list of only the words highlighted in the book

# Generate new empty list
new_list = []

# For each item in the results list add the first element (words we want to use)
for i in result_list:

    # Assign the word from the first element (position zero)
    i = i[0]

    # Add it to new list
    new_list.append(i)
new_list
```

```python
# Sort list alphabetically
new_list = sorted(new_list)
```

```python
# CSV files can be opened with excel or notepad
# Convert this alphabetically sorted list to CSV to work with in microsoft word

# import pandas
import pandas as pd

# Change list to dataframe
df = pd.DataFrame(new_list, columns= ["word"])

# Write dataframe to csv
#df.to_csv("poem_words.csv", index = False)
```
